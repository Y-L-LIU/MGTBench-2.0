```python
import json
import copy
```

```python
folder = './transfer'
```

```python
with open(folder + '/Moonshot_task3_transfer_results1.json', 'r') as f:
    res1 = json.load(f)

with open(folder + '/Moonshot_task3_transfer_results2.json', 'r') as f:
    res2 = json.load(f)
```

```python
for k in res1.keys():
    if len(res1[k]) == 0:
        print(k)
print()
for k in res2.keys():
    if len(res2[k]) == 0:
        print(k)
```

```python
res = copy.deepcopy(res1)
for k in res.keys():
    if len(res[k]) == 0:
        print(k)
        res[k] = res2[k]
```

```python
category = ['Physics', 'Medicine', 'Biology', 'Electrical_engineering', 'Computer_science', 
            'Literature', 'History', 'Education', 'Art', 'Law', 'Management', 'Philosophy', 
            'Economy', 'Math', 'Statistics', 'Chemistry']

print(len(res))
for c in category:
    print(c, len(res[c]))
```

```python
# save 
with open(folder + '/Moonshot_task3_transfer_results.json', 'w') as f:
    json.dump(res, f)
```
