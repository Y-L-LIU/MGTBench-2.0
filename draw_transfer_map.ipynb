```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json
```

```python
category = ['Physics', 'Medicine', 'Biology', 'Electrical_engineering', 'Computer_science', 
            'Literature', 'History', 'Education', 'Art', 'Law', 'Management', 'Philosophy', 
            'Economy', 'Math', 'Statistics', 'Chemistry']
```

```python
with open('transfer/Moonshot_task3_transfer_results.json') as f:
    data = json.load(f)
```

```python
print(data['Art'][0])
print(data['Art'][0]['Physics']['test']['f1'])
```

```python
# pairwise F1 score
num = len(category)
f1_scores = np.zeros((num, num))
for i in range(num):
    for j in range(num):
        f1_scores[i, j] = data[category[i]][j][category[j]]['test']['f1']
```

```python
# 创建DataFrame，用于表示F1 score矩阵
f1_matrix = pd.DataFrame(f1_scores, index=category, columns=category)

# 绘制热力图
plt.figure(figsize=(12, 10))
sns.heatmap(f1_matrix, annot=True,fmt=".3f", cmap="Blues", linewidths=0.7, square=True)

plt.title('Pairwise F1 Scores Between Categories', fontsize=16)

plt.show()
```
