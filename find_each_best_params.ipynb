### Find best hyperparams for distilbert/roberta and match/unmatch respectively


```python
# distilbert_unmatch = 'task3_ablation'
# distilbert_match = 'task3_ablation_match'
# roberta_unmatch = 'task3_ablation_roberta'
# roberta_match = 'task3_ablation_match_roberta'
distilbert_ratio2 = 'task3_ratio2'
distilbert_ratio2_match = 'task3_ratio2_match'
distilbert_ratio05 = 'task3_ratio0.5'
distilbert_ratio05_match = 'task3_ratio0.5_match'
# folders = [distilbert_unmatch, distilbert_match, roberta_unmatch, roberta_match]
folders = [distilbert_ratio2, distilbert_ratio05, distilbert_ratio2_match, distilbert_ratio05_match]
```

```python
category = ['Physics', 'Medicine', 'Biology', 'Electrical_engineering', 'Computer_science', 
            'Literature', 'History', 'Education', 'Art', 'Law', 'Management', 'Philosophy', 
            'Economy', 'Math', 'Statistics', 'Chemistry']
detectLLM = 'Moonshot'
for fold in folders[:]:
    try:
        with open(f'{fold}/results.json', 'r') as f:
            res = json.load(f)
    except FileNotFoundError:
        print(f'{fold} not found')
        continue
    # find the best hyperparameters for each (subject, detectLLM)
    best_hyperparams = {}
    for cat in category:
        best_hyperparams[cat] = {detectLLM: {}}
    
    for cat in category:
        subject_res_list = res[cat][detectLLM]
        best_f1 = 0
        best_hyper = None
        for subject_res in subject_res_list:
            if subject_res['test']['f1'] > best_f1:
                best_f1 = subject_res['test']['f1']
                best_hyper = subject_res
        model_name = 'roberta' if 'roberta' in fold else 'distilbert'
        best_hyperparams[cat][detectLLM] = {
            "model": model_name,
            "seed": best_hyper['seed'],
            "cut_length": best_hyper['cut_length'],
            "num_data": best_hyper['data_size'],
            "f1": round(best_hyper['test']['f1'], 4)
                }
    with open(f'{fold}/best_hyperparams.json', 'w') as f:
        json.dump(best_hyperparams, f)

        #  "model": "distilbert",
            # "seed": 3407,
            # "cut_length": 3000,
            # "num_data": 500,
            # "f1": 0.9259

    
```
